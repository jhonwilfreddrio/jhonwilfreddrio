## 1. Python Dice App

### Description:

A program that will roll a given number of dice of any number of
sides and sum the total. This program will continue to run until the user desires to quit.

```python
import random
name = input("Enter your name: ")
print(f"Hello {name}, Welcome to the Python Dice App")
dice_sides = []
dice_outcomes = []

while True:
      dice_side = int(input("\nHow many sides would you like on your dice: "))
      roll_count = int(input("How many dice would you like to roll: "))
      for dices in range(1, (dice_side + 1)):
          dice_sides.append(dices)
      print(f"\nYou rolled {roll_count} {dice_side} sided dice.\n")
      print("-----Results are as followed-----")
      for roll in range(roll_count):
        outcome = random.choice(dice_sides)
        dice_outcomes.append(outcome)
        print(outcome)
      all_outcome = sum(dice_outcomes)
      print(f"\nThe total value of your roll is {all_outcome}\n")
      restart = input("Would you like to roll again (y/n): \n")
      if restart == "n":
            print("Thank you for using the Python Dice App.")
            break
                
```

## 2. The Python Calculator App

### Description:

You are responsible for writing a program that simulates a calculator application that will take in
any two numbers, and a basic mathematical operation (addition, subtraction, multiplication,
division, or exponentiation), perform that operation, print a lexical statement of the operation,
and return a mathematical statement that describes the mathematical results. Upon completion,
your program will print out a history of all calculations performed including any error messages
that may have occurred such as division by zero.

```python
import random
import time
name = input("Enter your name: ")
print(f"Hello {name}, Welcome to the Python Calculator App")
time.sleep(1)
print("Enter two numbers and an operation and the desired operation will be performed.\n")
time.sleep(1)
records = []
multiplication = ["multiplication", "M", "m", "*"]
addition = ["addition", "A", "a", "+"]
subtraction = ["subtraction", "S", "s", "-"]
division = ["division", "D", "d", "/"]
exponentiation = ["exponentiation", "E", "e", "**", "^"]
while True:
    firstnum = float(input("\nEnter a number: "))
    secondnum = float(input("Enter a number: "))
    operation = input("\nEnter an operation (addition, subtraction, multiplication, division, or exponentiation): ")
    if operation in addition:
        answer = firstnum + secondnum
        print(f"The sum of {firstnum} and {secondnum} is {answer}.")
        records.append(f"{firstnum} + {secondnum} = {answer}")
    elif operation in subtraction:
        answer = firstnum - secondnum
        print(f"The difference of {firstnum} and {secondnum} is {answer}.")
        records.append(f"{firstnum} - {secondnum} = {answer}")
    elif operation in multiplication:
        answer = firstnum * secondnum
        print(f"The product of {firstnum} and {secondnum} is {answer}.")
        records.append(f"{firstnum} * {secondnum} = {answer}")
    elif operation in division:
        if secondnum == 0:
            records.append("DIV ERROR")
            print("You cannot divide by zero.")
        else:
            answer = firstnum / secondnum
            print(f"The quotient of {firstnum} and {secondnum} is {answer}.")
            records.append(f"{firstnum} / {secondnum} = {answer}")
    elif operation in exponentiation:
        answer = firstnum ** secondnum
        print(f"The quotient of {firstnum} and {secondnum} is {answer}.")
        records.append(f"{firstnum} ** {secondnum} = {answer}")
    else:
        print("That is not a valid operation. Try again.")
        records.append("OPP ERROR")
    restart = input("\nWould you like to run the program again (y/n): \n")
    if restart == "n":
        print("\nCalculation Summary:")
        for record in records:
            print(record)
```
